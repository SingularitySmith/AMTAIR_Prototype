```python
# first draft, can only handle one generational relations, ignores higher generations because of exact regex patterns

def parse_markdown_hierarchy(markdown_text):
    # Remove comments first
    markdown_text = re.sub(r'/\*.*?\*/', '', markdown_text, flags=re.DOTALL)

    # Initialize an empty DataFrame with all required columns
    df = pd.DataFrame(columns=[
        'Title', 'Description', 'Parents', 'Children', 'No_Parent', 'No_Children'
    ])

    # Split the markdown text into lines
    lines = markdown_text.split('\n')
    print(lines)
    # Regex pattern to extract text between angle brackets or square brackets
    title_pattern = r'[<\[](.+?)[>\]]'

    # Keep track of the last child title and its potential parents
    last_child_title = None
    current_parents = []
    nested_parents = []

    # Process each line
    for line in lines:
        # Trim the line
        stripped_line = line.strip()

        # Skip empty lines
        if not stripped_line:
            continue

        # Find titles in the current line
        titles = re.findall(title_pattern, line)

        # Check for nested parent levels
        if line.startswith('  +  ') or line.startswith('  -  '):
            # Two spaces after '+' indicates a nested parent
            nested_parent_match = re.findall(title_pattern, line)
            if nested_parent_match:
                nested_parents.append(nested_parent_match[0])

        # Check if this is a main title line (not starting with "  + ")
        # Ensure it starts with a title enclosed in < > or [ ] and doesn't start with "+ <" or "+ ["
        if (titles and
            re.match(r'^[<\[].*?[>\]]:', stripped_line) and
            not stripped_line.startswith('+ <') and
            not stripped_line.startswith('+ [') and
            not stripped_line.startswith('- <') and
            not stripped_line.startswith('- [')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'[<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''

                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': current_parents.copy(),  # Use a copy of current parents
                    'Children': [],
                    'No_Parent': len(current_parents) == 0,
                    'No_Children': True
                }

                # Add the row to the DataFrame
                df.loc[len(df)] = new_row

                # Update last child title
                last_child_title = title

                # Reset parents for the next main title
                current_parents = []
                nested_parents = []

        # Check if this is a parent line (starting with "  + ")
        elif titles and (stripped_line.startswith('+ <') or stripped_line.startswith('+ [') or stripped_line.startswith('  + ')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'\+ [<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''

                # Add this title to current parents
                current_parents.append(title)

                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': nested_parents.copy() if nested_parents else [],  # Use nested parents if available
                    'Children': [last_child_title] if last_child_title else [],
                    'No_Parent': len(nested_parents) == 0,
                    'No_Children': last_child_title is None
                }

                # Add the row to the DataFrame
                df.loc[len(df)] = new_row

                # If there was a last child, update its parents
                if last_child_title:
                    # Find the index of the last child row
                    last_child_index = df[df['Title'] == last_child_title].index[-1]

                    # Update parents to include both current and nested parents
                    combined_parents = current_parents.copy()
                    if nested_parents:
                        combined_parents.extend(nested_parents)

                    df.at[last_child_index, 'Parents'] = combined_parents
                    df.at[last_child_index, 'No_Parent'] = len(combined_parents) == 0

    return df


# Example usage
markdown_text = '''[AI_Affects_Power_Dynamics]: AI can shift or strengthen existing power dynamics between different actors.
  + [AI_State_State_Risk]: AI can disturb relationships between nation states, potentially leading to an "AI arms race."
  + [AI_State_Corporation_Risk]: The rise of tech corporations affects their relationships with states, creating power imbalances.
  +  [AI_State_Citizen_Risk]: AI surveillance technologies change the dynamics between states and citizens.
   /* => [AI_Existential_Risk_Factor]*/'''
#markdown_text = md_content

# Parse the markdown
result_df = parse_markdown_hierarchy(markdown_text)

# Display the DataFrame
result_df
```

```python
# function divided into several sub functions, to better handle
# in progress!

def parse_markdown_hierarchy(markdown_text):
    """Main function to parse markdown hierarchy into a DataFrame"""
    
    # Remove comments
    clean_text = remove_comments(markdown_text)
    
    # Extract all titles with their descriptions and indentation levels
    titles_info = extract_titles_info(clean_text)
    
    # Establish parent-child relationships
    titles_with_relations = establish_relationships(titles_info, clean_text)
    
    # Convert to DataFrame
    df = convert_to_dataframe(titles_with_relations)
    
    # Add No_Parent and No_Children columns
    df = add_no_parent_no_child_columns_to_df(df)
    
    return df

def remove_comments(markdown_text):
    """Remove comment blocks from markdown text"""
    return re.sub(r'/\*.*?\*/', '', markdown_text, flags=re.DOTALL)

def extract_titles_info(text):
    """Extract titles with their descriptions and indentation levels"""
    lines = text.split('\n')
    print(lines)
    titles_info = {}
    
    for line in lines:
        print(line)
        if not line.strip():
            continue
            
        title_match = re.search(r'[<\[](.+?)[>\]]', line)
        if not title_match:
            continue
            
        title = title_match.group(1)
        
        # Extract description
        title_pattern_in_line = r'[<\[]' + re.escape(title) + r'[>\]]:'
        description_match = re.search(title_pattern_in_line + r'\s*(.*)', line)
        description = description_match.group(1).strip() if description_match else ''
        
        indentation = 0
        if '+' in line:
            symbol_index = line.find('+')
            # Count spaces before the '+' symbol
            i = symbol_index - 1
            while i >= 0 and line[i] == ' ':
                indentation += 1
                i -= 1
        elif '-' in line:
            symbol_index = line.find('-')
            # Count spaces before the '-' symbol
            i = symbol_index - 1
            while i >= 0 and line[i] == ' ':
                indentation += 1
                i -= 1
        if indentation != 0:
            indentation = indentation-1
        # If neither symbol exists, indentation remains 0
        
        if title in titles_info:
            # Only update description if it's currently empty and we found a new one
            if not titles_info[title]['description'] and description:
                titles_info[title]['description'] = description
            # We might also need to update indentation if this is another occurrence
            if '+' in line and indentation > titles_info[title]['indentation']:
                titles_info[title]['indentation'] = indentation
        else:
            # First time seeing this title, create a new entry
            titles_info[title] = {
                'description': description,
                'indentation': indentation,
                'parents': [],
                'children': []
            }
            
    print(titles_info)
    return titles_info

def establish_relationships(titles_info, text):
    """Establish parent-child relationships between titles"""
    lines = text.split('\n')
    
    # Dictionary to store line numbers for each title
    title_lines = {}
    
    # Record line number for each title
    line_number = 0
    for line in lines:
        if not line.strip():
            line_number += 1
            continue
            
        title_match = re.search(r'[<\[](.+?)[>\]]', line)
        if not title_match:
            line_number += 1
            continue
            
        title = title_match.group(1)
        title_lines[title] = line_number
        line_number += 1
    
    # Sort titles by line number for sequential processing
    sorted_titles = sorted(titles_info.keys(), key=lambda t: title_lines.get(t, float('inf')))
    
    print("Sorted titles:")
    for title in sorted_titles:
        print(f"{title} - Line: {title_lines.get(title, 'Unknown')} - Indent: {titles_info[title]['indentation']}")
    
    # Process titles in the order they appear
    for i, title in enumerate(sorted_titles):
        current_indent = titles_info[title]['indentation']
        
        print(f"\nProcessing: {title} (indent: {current_indent})")
        
        # For non-indented titles (main titles)
        if current_indent == 0:
            print("  This is a main title, looking for parents...")
            # Look ahead for potential parents
            for j in range(i+1, len(sorted_titles)):
                next_title = sorted_titles[j]
                next_indent = titles_info[next_title]['indentation']
                
                print(f"    Checking {next_title} (indent: {next_indent})")
                
                # If we find another main title, stop looking
                if next_indent == 0:
                    print(f"    Found another main title, stopping parent search")
                    break
                    
                # If this is a direct parent (first level of indentation)
                if next_indent == 1:
                    print(f"    Found parent: {next_title}")
                    if next_title not in titles_info[title]['parents']:
                        titles_info[title]['parents'].append(next_title)
                    if title not in titles_info[next_title]['children']:
                        titles_info[next_title]['children'].append(title)
        
        # For indented titles
        else:
            print("  This is an indented title, looking for children...")
            # Look for the most recent title with one less indentation
            for j in range(i-1, -1, -1):
                prev_title = sorted_titles[j]
                prev_indent = titles_info[prev_title]['indentation']
                
                print(f"    Checking {prev_title} (indent: {prev_indent})")
                
                # If we find a title with exactly one less indentation
                if prev_indent == current_indent - 1:
                    print(f"    Found child: {prev_title}")
                    if title not in titles_info[prev_title]['parents']:
                        titles_info[prev_title]['parents'].append(title)
                    if prev_title not in titles_info[title]['children']:
                        titles_info[title]['children'].append(prev_title)
                    break
    
    print("\nFinal relationships:")
    for title, info in titles_info.items():
        print(f"{title} - Parents: {info['parents']} - Children: {info['children']}")
    
    return titles_info

def convert_to_dataframe(titles_info):
    """Convert the titles information dictionary to a pandas DataFrame"""
    df = pd.DataFrame(columns=['Title', 'Description', 'Parents', 'Children'])
    
    for title, info in titles_info.items():
        df.loc[len(df)] = {
            'Title': title,
            'Description': info['description'],
            'Parents': info['parents'],
            'Children': info['children']
        }
    
    return df

def add_no_parent_no_child_columns_to_df(dataframe):
    """Add No_Parent and No_Children boolean columns to the DataFrame"""
    no_parent = []
    no_children = []
    
    for _, row in dataframe.iterrows():
        no_parent.append(not row['Parents'])
        no_children.append(not row['Children'])
    
    dataframe['No_Parent'] = no_parent
    dataframe['No_Children'] = no_children
    
    return dataframe
```

```python
# function divided into several sub functions, to better handle, added functionality to add metadata
# in progress!

def parse_markdown_hierarchy(markdown_text):
    """Main function to parse markdown hierarchy into a DataFrame"""
    
    # Remove comments
    clean_text = remove_comments(markdown_text)
    
    # Extract all titles with their descriptions and indentation levels
    titles_info = extract_titles_info(clean_text)
    
    # Establish parent-child relationships
    titles_with_relations = establish_relationships(titles_info, clean_text)
    
    # Convert to DataFrame
    df = convert_to_dataframe(titles_with_relations)
    
    # Add No_Parent and No_Children columns
    df = add_no_parent_no_child_columns_to_df(df)
    
    return df

def remove_comments(markdown_text):
    """Remove comment blocks from markdown text"""
    return re.sub(r'/\*.*?\*/', '', markdown_text, flags=re.DOTALL)

def extract_titles_info(text):
    """Extract titles with their descriptions and indentation levels"""
    lines = text.split('\n')
    print(lines)
    titles_info = {}
    
    for line in lines:
        print(line)
        if not line.strip():
            continue
            
        title_match = re.search(r'[<\[](.+?)[>\]]', line)
        if not title_match:
            continue
            
        title = title_match.group(1)
        
        # Extract description and metadata
        title_pattern_in_line = r'[<\[]' + re.escape(title) + r'[>\]]:'
        description_match = re.search(title_pattern_in_line + r'\s*(.*)', line)

        if description_match:
            full_text = description_match.group(1).strip()
            
            # Check if description contains a "{" to not include metadata in description
            if "{" in full_text:
                # Split at the first "{"
                split_index = full_text.find("{")
                description = full_text[:split_index].strip()
                metadata = full_text[split_index:].strip()
            else:
                # Keep the entire description and no metadata
                description = full_text
                metadata = ''
        else:
            description = ''
        
        indentation = 0
        if '+' in line:
            symbol_index = line.find('+')
            # Count spaces before the '+' symbol
            i = symbol_index - 1
            while i >= 0 and line[i] == ' ':
                indentation += 1
                i -= 1
        elif '-' in line:
            symbol_index = line.find('-')
            # Count spaces before the '-' symbol
            i = symbol_index - 1
            while i >= 0 and line[i] == ' ':
                indentation += 1
                i -= 1
        if indentation != 0:
            indentation = indentation-1
        # If neither symbol exists, indentation remains 0
        
        if title in titles_info:
            # Only update description if it's currently empty and we found a new one
            if not titles_info[title]['description'] and description:
                titles_info[title]['description'] = description
            # We might also need to update indentation if this is another occurrence
            if '+' in line and indentation > titles_info[title]['indentation']:
                titles_info[title]['indentation'] = indentation
        else:
            # First time seeing this title, create a new entry
            titles_info[title] = {
                'description': description,
                'indentation': indentation,
                'parents': [],
                'children': [],
                'metadata': metadata
            }
            
    print(titles_info)
    return titles_info

def establish_relationships(titles_info, text):
    """Establish parent-child relationships between titles"""
    lines = text.split('\n')
    
    # Dictionary to store line numbers for each title
    title_lines = {}
    
    # Record line number for each title
    line_number = 0
    for line in lines:
        if not line.strip():
            line_number += 1
            continue
            
        title_match = re.search(r'[<\[](.+?)[>\]]', line)
        if not title_match:
            line_number += 1
            continue
            
        title = title_match.group(1)
        title_lines[title] = line_number
        line_number += 1
    
    # Sort titles by line number for sequential processing
    sorted_titles = sorted(titles_info.keys(), key=lambda t: title_lines.get(t, float('inf')))
    
    print("Sorted titles:")
    for title in sorted_titles:
        print(f"{title} - Line: {title_lines.get(title, 'Unknown')} - Indent: {titles_info[title]['indentation']}")
    
    # Process titles in the order they appear
    for i, title in enumerate(sorted_titles):
        current_indent = titles_info[title]['indentation']
        
        print(f"\nProcessing: {title} (indent: {current_indent})")
        
        # For non-indented titles (main titles)
        if current_indent == 0:
            print("  This is a main title, looking for parents...")
            # Look ahead for potential parents
            for j in range(i+1, len(sorted_titles)):
                next_title = sorted_titles[j]
                next_indent = titles_info[next_title]['indentation']
                
                print(f"    Checking {next_title} (indent: {next_indent})")
                
                # If we find another main title, stop looking
                if next_indent == 0:
                    print(f"    Found another main title, stopping parent search")
                    break
                    
                # If this is a direct parent (first level of indentation)
                if next_indent == 1:
                    print(f"    Found parent: {next_title}")
                    if next_title not in titles_info[title]['parents']:
                        titles_info[title]['parents'].append(next_title)
                    if title not in titles_info[next_title]['children']:
                        titles_info[next_title]['children'].append(title)
        
        # For indented titles
        else:
            print("  This is an indented title, looking for children...")
            # Look for the most recent title with one less indentation
            for j in range(i-1, -1, -1):
                prev_title = sorted_titles[j]
                prev_indent = titles_info[prev_title]['indentation']
                
                print(f"    Checking {prev_title} (indent: {prev_indent})")
                
                # If we find a title with exactly one less indentation
                if prev_indent == current_indent - 1:
                    print(f"    Found child: {prev_title}")
                    if title not in titles_info[prev_title]['parents']:
                        titles_info[prev_title]['parents'].append(title)
                    if prev_title not in titles_info[title]['children']:
                        titles_info[title]['children'].append(prev_title)
                    break
    
    print("\nFinal relationships:")
    for title, info in titles_info.items():
        print(f"{title} - Parents: {info['parents']} - Children: {info['children']}")
    
    return titles_info

def convert_to_dataframe(titles_info):
    """Convert the titles information dictionary to a pandas DataFrame"""
    df = pd.DataFrame(columns=['Title', 'Description', 'Parents', 'Children', 'instantiations', 'priors', 'posteriors'])
    
    for title, info in titles_info.items():
        # Parse the metadata JSON string into a Python dictionary
        if 'metadata' in info and info['metadata']:
            try:
                jsonMetadata = json.loads(info['metadata'])
                
                # Create the row dictionary with basic fields
                row = {
                    'Title': title,
                    'Description': info.get('description', ''),
                    'Parents': info.get('parents', []),
                    'Children': info.get('children', []),
                    # Extract specific metadata fields, defaulting to empty if not present
                    'instantiations': jsonMetadata.get('instantiations', []),
                    'priors': jsonMetadata.get('priors', {}),
                    'posteriors': jsonMetadata.get('posteriors', {})
                }
            except json.JSONDecodeError:
                # Handle case where metadata isn't valid JSON
                row = {
                    'Title': title,
                    'Description': info.get('description', ''),
                    'Parents': info.get('parents', []),
                    'Children': info.get('children', []),
                    'instantiations': [],
                    'priors': {},
                    'posteriors': {}
                }
        else:
            # Handle case where metadata field doesn't exist
            row = {
                'Title': title,
                'Description': info.get('description', ''),
                'Parents': info.get('parents', []),
                'Children': info.get('children', []),
                'instantiations': [],
                'priors': {},
                'posteriors': {}
            }
        
        # Add the row to the DataFrame
        df.loc[len(df)] = row
    
    return df

def add_no_parent_no_child_columns_to_df(dataframe):
    """Add No_Parent and No_Children boolean columns to the DataFrame"""
    no_parent = []
    no_children = []
    
    for _, row in dataframe.iterrows():
        no_parent.append(not row['Parents'])
        no_children.append(not row['Children'])
    
    dataframe['No_Parent'] = no_parent
    dataframe['No_Children'] = no_children
    
    return dataframe
```
